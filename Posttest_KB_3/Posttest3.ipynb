Import Library

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load Dataset

```python
df = pd.read_csv("TravelInsurancePrediction.csv")
```

Drop Kolom Tidak Bernama

```python
df = df.drop(['Unnamed: 0'], axis=1)
```

Cek Data

```python
df.head()
```

## GRAFIK BATANG
Membandingkan jumlah label Employment Type berdasarkan setiap label pada kolom Travel Insurance

```python
sns.countplot(x=df['TravelInsurance'], hue = df['Employment Type'], edgecolor = 'black')
plt.title("Jumlah Employment Type Pada Kolom Travel Insurance", weight='bold', fontsize = 14)
plt.ylabel("Jumlah")
plt.xlabel("Travel Insurance")
plt.show()
```

### KESIMPULAN

Hal yang didapat :
 1. Pada kolom Travel Insurance dengan nilai 0 dan employment type berlabel goverment sector sebanyak kurang lebih 400, sedangkan yang berlabel private sector/self employed lebih dari 800
 2. Pada kolom Travel Insurance dengan nilai 1 dan employment type berlabel goverment sector sebanyak kurang lebih 100, sedangkan yang berlabel private sector/self employed lebih dari 500

Dari hal berikut dapat disimpulkan bahwa :
1. jumlah karyawan kedua tipe yang memiliki travel insurance lebih sedikit dibanding dengan yang tidak memiliki travel insurance.
2. jumlah karyawan yang memiliki travel insurance dengan tipe goverment sector lebih sedikit dari private sector.
3. jumlah karyawan yang tidak memiliki travel insurance dengan tipe goverment sector lebih sedikit dari private sector.
4. kedua tipe karyawan cenderung tidak memiliki travel insurance, tetapi untuk rasio yang lebih banyak memiliki travel insurance ialah tipe private sector dengan perbandingan 5:8, sedangkan untuk government sector 1:4 atau 2:8


## GRAFIK PIE

```python
df_jumlah_employ = df['Employment Type'].value_counts()
plt.pie(
    df_jumlah_employ,
    labels = df_jumlah_employ.index,
    autopct = "%1.2f%%",
    wedgeprops = {'edgecolor':'black'}
    )
plt.title("Perbandingan Jumlah Employment Type", weight='bold', fontsize = 14)
plt.show()
```

### KESIMPULAN
Hal yang didapat:
1. Jumlah yang merupakan tipe government sector ialah sekitar 28.69% dari jumlah keseluruhan
2. Jumlah yang merupakan tipe private sector/self employed sekitar 71.31% dari jumlah keseluruhan

Dapat disimpulkan:
1. yang merupakan government sector lebih sedikit dari private sector dengan perbandingan sekitar 30% untuk government sector dan 70% untuk private sector

## GRAFIK HEATMAP

```python
corr = df.corr(numeric_only = True)
sns.heatmap(corr, annot=True, cmap='Oranges', center = 1)
plt.title('Korelasi')
plt.show()
```

### KESIMPULAN

Hal yang didapat:
1. Korelasi dengan pengaruh tertinggi berada pada Annual Income dengan Travel Insurance dengan nilai 0.4
2. Banyak kolom yang hampir tidak ber korelasi dengan nilai yang sangat kecil mendekati 0

Dapat disimpulkan:
1. Banyak karyawan yang jika annual incomenya tinggi maka ia cenderung akan membeli travel insurance begitu juga sebaliknya dengan kemungkinan 40% untuk membeli travel insurance
