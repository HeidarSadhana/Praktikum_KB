#### Import Pandas dan buat dataframe dengan membaca dataset 

```python
import pandas as pd
dataframe = pd.read_csv("TravelInsurancePrediction.csv")
```

#### Drop kolom index dari file csv

```python
df = dataframe.drop(['test'], axis=1)
```

#### Membuat dictionary informasi dengan key jumlah record, jumlah atribut, dan jumlah atribut bertipe angka

```python
informasi = {
    'Jumlah Record': len(df),
    'Jumlah Atribut': len(df.columns),
    'Atribut Tipe Angka': len(df.select_dtypes('int64','float64').columns),
}
```

#### Membuat dictionary untuk menampilkan setiap kolom yang ada dan tipe datanya

```python
info_nama = []
for i in df.columns.tolist():
    info_nama.append(
        {
            "Nama Atribut" : i, 
            "Tipe Atribut" : df[i].dtype
        }
    )
```

#### Menampilkan isi dari dictionary yang sudah dibuat menggunakan for loop

```python
for key, value in informasi.items():
    print(f"{key} : {value}")
for dik in info_nama:
    print("-------------------------------")
    for key, value in dik.items():
        print(f"{key} : {value}")
```
